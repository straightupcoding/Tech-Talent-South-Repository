### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = pd.read_csv(r"C:\Users\aguan\Documents\company_sales_data.csv")

listOfMonths = data ['month_number'].tolist()
faceCreamSalesData = data ['facecream'].tolist()
faceWashSalesData = data ['facewash'].tolist()
toothPasteSalesData = data ['toothpaste'].tolist()
bathingSoapSalesData = data ['bathingsoap'].tolist()
shampooSalesData = data ['shampoo'].tolist()
moisturizerSalesData = data ['moisturizer'].tolist()

plt.plot(listOfMonths, faceCreamSalesData,   label = 'Face cream Sales Data')
plt.plot(listOfMonths, faceWashSalesData,   label = 'Face Wash Sales Data', marker ='o')
plt.plot(listOfMonths, toothPasteSalesData, label = 'ToothPaste Sales Data')
plt.plot(listOfMonths, bathingSoapSalesData, label = 'Bathing Soap Sales Data')
plt.plot(listOfMonths, shampooSalesData, label = 'Shampoo Sales Data')
plt.plot(listOfMonths, moisturizerSalesData, label = 'Moisturizer Sales Data')

plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(loc='lower right')
plt.xticks(listOfMonths)
plt.show
````

### Exercise 4.

Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should .

````python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(r"C:\Users\aguan\Documents\company_sales_data.csv")
listOfMonths = data ['month_number'].tolist()
salesDataForToothepaste = data ['toothpaste'].tolist()
plt.scatter(listOfMonths, salesDataForToothepaste, label = 'Sales data - Tooth Paste')
plt.xlabel('Month #')
plt.ylabel('Units Sold')
plt.legend(loc='upper left')
plt.title('Sales Data - Toothepaste')
plt.xticks(listOfMonths)
plt.grid(True, linewidth=1, linestyle="-")
plt.show()
````

### Exercise 5.

Read face cream and face wash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

````python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\aguan\Documents\company_sales_data.csv")
listOfMonths = data ['month_number'].tolist()
faceCreamSales = data ['facecream'].tolist()
faceWashSales = data ['facewash'].tolist()

plt.bar([a-.5 for a in listOfMonths], faceCreamSales, width=0.5, label = 'Sales Data - Face Cream', align='edge')
plt.bar([a+.5 for a in listOfMonths], faceWashSales, width= -0.5, label = 'Sales Data - Face Wash', align='edge')
plt.xlabel('Month #')
plt.ylabel('Sales Units')
plt.legend(loc='lower right')
plt.title('Sales Data')

plt.grid(True, linewidth = 1, linestyle = "--")
plt.title('Sales Data - FaceWash & FaceCream')
plt.show()
````
