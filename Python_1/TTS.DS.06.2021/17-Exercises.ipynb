### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = pd.read_csv(r"C:\Users\aguan\Documents\company_sales_data.csv")

listOfMonths = data ['month_number'].tolist()
faceCreamSalesData = data ['facecream'].tolist()
faceWashSalesData = data ['facewash'].tolist()
toothPasteSalesData = data ['toothpaste'].tolist()
bathingSoapSalesData = data ['bathingsoap'].tolist()
shampooSalesData = data ['shampoo'].tolist()
moisturizerSalesData = data ['moisturizer'].tolist()

plt.plot(listOfMonths, faceCreamSalesData,   label = 'Face cream Sales Data')
plt.plot(listOfMonths, faceWashSalesData,   label = 'Face Wash Sales Data')
plt.plot(listOfMonths, toothPasteSalesData, label = 'ToothPaste Sales Data')
plt.plot(listOfMonths, bathingSoapSalesData, label = 'ToothPaste Sales Data')
plt.plot(listOfMonths, shampooSalesData, label = 'ToothPaste Sales Data')
plt.plot(listOfMonths, moisturizerSalesData, label = 'ToothPaste Sales Data')

plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(loc='lower right')
plt.xticks(listOfMonths)
plt.show
````
